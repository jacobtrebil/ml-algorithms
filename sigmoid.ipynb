```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
print(sigmoid(0.5))
```

# Sigmoid Function as a Lambda Function

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
```

# Sigmoid with Scipy

```python
from scipy.special import expit
print(expit(0.5))
```

# Applying the Sigmoid Function to Numpy Arrays

```python
arr = np.array([-2, -1.5, -1, 0, 1, 1.5, 2])
print(sigmoid(arr))
```

# Applying the Sigmoid Function to Python Lists

```python
values = [-2, -1.5, -1, 0, 1, 1.5, 2]
print([sigmoid(value) for value in values])
```

# Plotting the Sigmoid Function with Matplotlib

```python
values = np.arange(-10, 10, 0.1)

plt.plot(values, sigmoid(values))
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function')
plt.show()
```
